```python
import graphlab
products = graphlab.SFrame('amazon_baby.gl/')
```

```python
products.head()
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
products['word_count'] = graphlab.text_analytics.count_words(products['review'])
```

```python
def awesome_count(x):
    k=10
    if selected_words[k] in x:
        return x[selected_words[k]]
    else:
        return 0
```

```python
products[selected_words[0]] = products['word_count'].apply(awesome_count)
print selected_words[0] +" " + str(sum(products[selected_words[0]]))
    
    
```

```python
j=10
products[selected_words[j]] = products['word_count'].apply(awesome_count)
print selected_words[j] +" " + str(sum(products[selected_words[j]]))
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
products = products[products['rating'] != 3]
```

```python
products['sentiment'] = products['rating'] >=4
```

```python
products.head()
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
selected_word_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=selected_words,
                                                     validation_set=test_data)
```

```python
selected_word_model['coefficients']
```

```python
reviews = selected_word_model['coefficients'].sort('value', ascending=False)
```
